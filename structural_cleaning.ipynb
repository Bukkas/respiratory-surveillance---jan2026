# Notebook 01 — Structural Cleaning (Suveillance Dataset)
This notebook performs the initial data profiling: **Structural Cleaning**.

Here, we load the raw surveillance data and identify:
1. Column names to be standardized
2. Inconsistent data types
3. Structural duplicates
4. Date-like columns to be converted

This step ensures the dataset is fit for further cleaning (categorical, numerical)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/raw/outbreak_surveillance_data.csv')
clean_df = df
df.head()
```

## Initial Dataset Overview
Before making any changes, we review:

1. Dataset shape
2. Column names
3. Data types
4. Number of missing values
5. A quick visual feel

```python
print('Dataset Shape:', clean_df.shape)
print('\nColumn Names:', clean_df.columns.tolist())
print('\nData Types:\n', clean_df.dtypes)
print('\nMissing Values:\n', clean_df.isna().sum())
```

### Column Names Standardization

```python
clean_df.columns = (
    clean_df.columns.astype(str)
    .str.lower()
    .str.strip()
    .str.replace(r'[ -]+', '_', regex = True)
)
clean_df.head()
```

### Data Types Fixing

**Date-like column**

```python
clean_df['collection_date'] = pd.to_datetime(clean_df['collection_date'], errors = 'coerce')
```

**Categorical Columns**

```python
cat_col = [
    'location',
    'pathogen'
]

for col in cat_col:
    df[col] = df[col].astype('category')
```

### Checking and Removing Duplicate Records

```python
before = clean_df.shape[0]
clean_df = clean_df.drop_duplicates()
after = clean_df.shape[0]

print(f'Removed {before - after} duplicate rows.')
```
